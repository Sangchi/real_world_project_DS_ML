```python
x=dataset.iloc[:,1:5].values  #it will convert dataframe into numpy array(values)
y=dataset.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelBinarizer
lb=LabelBinarizer()
y=lb.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=100)
classifier.fit(x_train,y_train)
```

```python
y_pridict=classifier.predict(x_test)
y_pridict
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(np.argmax(y_pridict,axis=1),np.argmax(y_test,axis=1))


```

```python
cm
```

```python
accc=accuracy_score(np.argmax(y_pridict,axis=1),np.argmax(y_test,axis=1))
```
